```python
last_name = input("Яке ваше прізвище?")
points = int(input("Яка ваша кількість балів?"))

result = None
if (points < 60):
    orange = 'незадовільно'
    result = print(f"Студент {last_name} набрав {points} балів та отримав оцінку {orange}")
elif (points <= 74):
    orange = 'задовільно'
    result= print(f"Студент {last_name} набрав {points} балів та отримав оцінку {orange}")
elif (points <= 94):
    orange = 'добре'
    result= print(f"Студент {last_name} набрав {points} балів та отримав оцінку {orange}")
elif (points <= 100):
    orange = 'відмінно'
    result= print(f"Студент {last_name} набрав {points} балів та отримав оцінку {orange}")
else :
    print('ERROR')
```

```python
#приймає перші дві літри номеру транспортного засобу
cod_region = input('Ведіть перші дві літери номеру транспортного засобу:')

region = None
match cod_region:    #визначає область реєстрації авто 
    case "AA" | "KA":
        region = "місто Київ"    
    case "AB":
        region = "Вінницька область"   
    case "AE":
        region = "Дніпропетровська область"
    case "AH":
        region = "Донецька область"
    case _ :
        region = "інша область"
   
       
finish = f"{cod_region} - {region}"   
print(finish)
```
